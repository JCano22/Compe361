```python
u = "abcd"
z = 'ijkl'

print(u)
print(u[1]) #prints char at index 1
print(u[1:3]) #extracts a part of the string in the range of 1-3
print(u[1:4])

h = u[1:4] #new string with portion of "u"
print(h + " is the new string.")

print(z + " this is 'z' before.")
z = u #you don't copy the object, you copy the address, address to 'ijkl' is lost

print(z + " this is 'z' after.")
```

## String functions

-Length of a String:

`print(len(string)) //prints lenght of strenght`

-Accessing a character:

`print(string[4]) //prints the char at index 4`

-Accessing a subset of chars:

`print(string[1:5]) //prints characters from index 1 up to (but not including) index 5`

`print(string[:5]) //prints characters from the beginning up to index 5 (non-inclusive)`

`print(string[2:]) //prints characters from index 2 to the end of the string`

-Splitting string using `split()`:

`print(string.split(' ')) //splits string at spaces, if string is 'Hello there', result is 'Hello', 'there'`

-Counting strings using `count()`:

`print(string.count(' ')) //counts number of times a space is found in the string`

-Replacing strings using `replace()`:

`print(string.replace('word1', 'word2')) //replaces the ocurrence of word1 with word2`

-Find substrings using `find()`:

`print(string.find('find')) //finds substring in the parenthesis, returns index`

-Additional functions include:

`string.upper() //converts chars to uppercase`

`string.lower() //converts chars to lowercase`

`string.title() //capitalizes first char of each string`

### If stmnts

Example:

```python
p = 2
q = 3

if p > q:
    print("p is larger")
else:
    print("p is not larger")
```

- `elif` -> equivalent else if in C++

```python
if 5 > 6:
    print('yes')
elif 5 > 7:
    print('yes')
else:
    print('no')
```

### Match/Case stmnts

-Similar to switch statements in C++:

```python
num = 2

match num:
    case 1:
        print('1')
    case 2:
        print('yes, it is 2')
    case 3: 
        print('3')
```

### Ternary Operator in Python
-Similar to ternary in C++:

```python
n = 4
res = "Even" if n % 2 == 0 else "Odd"
print(res)
```

### Logical Operators

- `and` -> Returns True if both left and right are true
- `or` -> Returns True if either the left or the right is true
- `not` -> Returns true if the value being tested is False

```python
first = True
second = False
third = True

print(first or second)
print(first and third)
print(not second) # same as !second in C++
```

### ***While loops***
-Similar to C++: 

`while <condition>:
    do something`

```python
# Greatest common divisor
p = 75
q = 45

# while they are not equal, will subtract smallest from largest until numbers are equal indicating GCD

while p != q:
    if p > q:
        p -= q
    else:
        q -= p

print(p)
```

-For loops and While loops can have an `else: ` section following the loop.

-The `else: ` block only runs after the while loop has been exited normally and not by `break`

```python
# Checking for prime numbers

p = 1001
j = 2

# while j is less than p, it will check if it is divisible by j, if it is 'Not prime' and break the loop, otherwise it will continue to check unitl they are equal, and then the `else:` clause activates to print 'Prime'

while j < p:
    if p % j == 0:
        print('Not prime.')
        break
    j += 1
else:
    print('Prime.')
```

### ***For loops***

-Loop through a `range`.

-Parenthesis have `(start, stop, step)`

-If no step is included, loop will increment by 1

Example:

```python
for j in range(5, 13 + 1, 2):
    print(j)
```

Example using array:

```python
u = [9, 3, 7, 4, 8] # python list

# z is the loop variable that represents each element of the list
for z in u:
    print(z)

# using index
for j in range(0, len(u)):
    print(u[j])
```
