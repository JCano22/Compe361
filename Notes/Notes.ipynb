## **Strings**
- Strings are immutable in Python

`u = "abcd" //u is a pointer to "abcd"`

`z = 'ijkl' //single quotations also valid`


```python
u = "abcd"
z = 'ijkl'

print(u)
print(u[1]) #prints char at index 1
print(u[1:3]) #extracts a part of the string in the range of 1-3
print(u[1:4])

h = u[1:4] #new string with portion of "u"
print(h + " is the new string.")

print(z + " this is 'z' before.")
z = u #you don't copy the object, you copy the address, address to 'ijkl' is lost

print(z + " this is 'z' after.")
```

- It is not possible to replace/change strings.

`u = "abcd" //assigning u`

`u[1] = 'q' //not allowed`



### String concatenation: 

```python
p = "Poway"
q = 'California'

r = p + q # can be applied to numbers and to strings as well

print(r)

r = p + ", " + q #can also add other chars
print(r)
```

What happens when you try to concatenate an int with a string?:


```python
p = 1000
q = "527"

#r = p + q
#print(r) # will cause TypeError "unsupporter operand type"
```

Would need to cast one to either 'str' or 'int':

```python
p = 1000
q = "527"

r = str(p) + q
print(r)

#could also cast to 'int'
r = p + int(q)
print(r)
```

Next what happens with type 'float'?:

```python
p = 1000
q = "527.9"

r = p + float(q) #casting to float
print(r)
```

* look into `StringIO()`

StringIO() needs header: `from io import StringIO`





```python
from io import StringIO

u = StringIO()

u.write("abcd")
u.write("ijkl")
u.write("mnop")

r = u.getvalue()
u.close()
u = None

print(r)
```
